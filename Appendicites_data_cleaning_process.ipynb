FILLING MISSING VALUES IN NON NUMERIC COLUMNS

We now replace the missing non numeric values .


1.   Identify the non numeric columns .There are two types of non numeric values .categorical and boolean values
2.   we create a dictionary to store the predefined values specifically in non numeric columns .
3.we use For loop for a iterating each column of each non numeric set .
4. we check whether the curruent col is diagnosis.we replace the unique values and null values with no appendicites .
5.Now we go for  binary or categorical columns  such as sex, management ,severity







6. The missing values are filled with the mode value .if there is no mode then it is filled with a no assuming absence of the symptom or feature.
7.   WE change it with mode so that it could maintain real world relavance .It also maintance consistency.(if men were impacted before the fill of null values ,men will be seen mostly affected after filling null values ).






8. Stool has specific values(normal", "diarrhea", "constipation".) ,we handle it diffrent .we replace the null values with the mode ,ie normal
9. unknown is used for specific columns to avoid patterns that might lead to misleading predictions.
10. We save the changes in the app_data_updated.csv  file .



```python


# Step 4: Replace NaN values in non-numeric columns with appropriate values
non_numeric_columns = df.select_dtypes(include=['object', 'bool']).columns

# Dictionary to store replacement values for non-numeric columns
replacement_values = {}

# For each non-numeric column, determine the replacement value
for col in non_numeric_columns:
    if col == 'Diagnosis':
        # Already handled above, use 'no appendicitis' for nulls
        replacement_values[col] = 'no appendicitis'
    elif col in ['Sex', 'Management', 'Severity', 'Appendix_on_US', 'Migratory_Pain',
                 'Lower_Right_Abd_Pain', 'Contralateral_Rebound_Tenderness', 'Coughing_Pain',
                 'Nausea', 'Loss_of_Appetite', 'Peritonitis', 'Psoas_Sign',
                 'Ipsilateral_Rebound_Tenderness', 'US_Performed', 'Free_Fluids',
                 'Target_Sign', 'Appendicolith', 'Perforation', 'Surrounding_Tissue_Reaction',
                 'Appendicular_Abscess', 'Pathological_Lymph_Nodes', 'Bowel_Wall_Thickening',
                 'Conglomerate_of_Bowel_Loops', 'Ileus', 'Coprostasis', 'Meteorism', 'Enteritis']:
        # For binary/categorical columns, use the mode (most frequent value)
        mode_value = df[col].mode().iloc[0] if not df[col].mode().empty else 'no'
        replacement_values[col] = mode_value
    elif col == 'Stool':
        # Stool has specific values: normal, diarrhea, constipation
        mode_value = df[col].mode().iloc[0] if not df[col].mode().empty else 'normal'
        replacement_values[col] = mode_value
    elif col in ['Diagnosis_Presumptive', 'Appendix_Wall_Layers', 'Abscess_Location',
                 'Lymph_Nodes_Location', 'Gynecological_Findings']:
        # For free-text or complex columns, use 'unknown' or 'none' as a default
        replacement_values[col] = 'unknown'
    else:
        # Default for any other non-numeric column
        replacement_values[col] = 'unknown'

# Replace NaN values in non-numeric columns
for col, value in replacement_values.items():
    df[col] = df[col].fillna(value)

# Save the updated DataFrame to a new CSV
df.to_csv('app_data_updated.csv', index=False)

print(" null values in non-numeric columns "
      "replaced with contextually appropriate values. Saved to 'app_data_updated.csv'.")
```

 STANDARDIZING CONTINOUS VARIABLES .


1.   Import pandas ,a data manipulation library
2.   import sklearn.preprocessing ,a module from scikit learn library .This library is used from numerical computations .StandardScalar is  a feature scaling technique from the above mentioned module .
3. Scaling is essential to prevent large scale variables dominating the smaller ones .
4. We load the data set .we read the app_data_updated.csv  file under pandas data frame for manupilation .
5. After reading ,we select the continous variables (numeric variables )that requires standardization.
6. filter the columns that are only  present in the app_data_updated.csv for processing .
7. Replace unknown values to Nan .non numeric values do not interfere with numeric values.
8. the nan values are replaced with mean of each column .
9. standard scalar does tranformations in such a way that the features have mean =0 and deviation as 1.This tranformation is applied only to continous variables .
10. identify categorical ,binary variable.
11. Mege the standarized numeric data with non standarized categorical data.
12. save the processed data in a new file standardized_data.csv




```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load the data
df = pd.read_csv('app_data_updated.csv')

# List of continuous variables to standardize
continuous_vars = [
    'Age', 'BMI', 'Height', 'Weight', 'Length_of_Stay', 'Appendix_Diameter',
    'Body_Temperature', 'WBC_Count', 'Neutrophil_Percentage', 'RBC_Count',
    'Hemoglobin', 'RDW', 'Thrombocyte_Count', 'CRP'
]

# Filter only columns that exist in the dataset
continuous_vars = [col for col in continuous_vars if col in df.columns]

# Handle missing values (replace 'unknown' or empty strings with NaN, then fill with mean)
df[continuous_vars] = df[continuous_vars].replace('unknown', pd.NA)
df[continuous_vars] = df[continuous_vars].apply(pd.to_numeric, errors='coerce')
df[continuous_vars] = df[continuous_vars].fillna(df[continuous_vars].mean())

# Standardize the continuous variables
scaler = StandardScaler()
standardized_data = pd.DataFrame(
    scaler.fit_transform(df[continuous_vars]),
    columns=continuous_vars
)

# Combine standardized continuous variables with non-continuous columns
non_continuous_vars = [col for col in df.columns if col not in continuous_vars]
combined_data = pd.concat([standardized_data, df[non_continuous_vars].reset_index(drop=True)], axis=1)

# Save to a new CSV file
combined_data.to_csv('standardized_data.csv', index=False)

# Show the first five rows as output
print("First five rows of the combined standardized dataset:")
print(combined_data.head())
```

BORUTA FEATURE SELECTION USING RANDOM FOREST CLASSSIFIER TO IDENTIFY MOST IMP FEATURES OF DYSURIA



1.   Import necessary libraries . pandas for data manipulation,numpy for numerical operations,
2.   Random classifier an assemble ml alogorithm from scikit-learn.
3.  Boruta used to calculate feature importance .
4.   Load the app_data_updated file in the panadas dataframe.
5.  define the target ,ie Dysuria.
6. Exclude the target column .
7. Use boruta on the  other columns expect thye target column .
8. Do encoding for categorical data ,ie give a value of 0,1
9. We fill the empty numericals columns with its median.
10. Do the above  encodings for target column
11. Use random forestclassifier for feature selection.Let the shuffling take place for first 10 ,5 to prevent overfitting .
12. boruto uses mutiple iteration using random forest ,n-jobs = number of cpu cores used for numeric calculations.class weight balanced=Used so that the columns with less values are treated equally like the once with max values.random state =The randomness pattern changes for each number.more the value ,more the uniqueness .The probability of getting the same pattern is low if more value like 65.
13. Select the encoded values and target values.
14. select the features to be selected and makes the list of them.
15. Run the boruto algorith on encoded and target features.
16.identify the importantant features after boruto as boruta_selector.support.
17. print the imp features
[] indicates no features were selected.





```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy

# Load the dataset
df = pd.read_csv("/content/app_data_updated .csv")

# Define target and features
target = "Dysuria"
features = df.columns[df.columns != target]  # Exclude target column

# Convert categorical features to numerical (if needed)
df_encoded = pd.get_dummies(df[features])

# Handle missing values in features
df_encoded.fillna(df_encoded.median(), inplace=True)

# Convert target variable to binary values and handle NaNs
df[target] = df[target].map({"yes": 1, "no": 0})
df[target].fillna(0, inplace=True) # Fill NaN values with 0

# Initialize the RandomForest model
rf = RandomForestClassifier(n_jobs=-1, class_weight="balanced", max_depth=10, random_state=42)

# Initialize Boruta
boruta_selector = BorutaPy(rf, n_estimators="auto", verbose=3, random_state=42)

# Fit Boruta to the dataset
boruta_selector.fit(df_encoded.values, df[target].values)

# Get selected features
selected_features = df_encoded.columns[boruta_selector.support_].tolist()

print("Selected features for Dysuria:", selected_features)
```

FEATURE SELECTION USING BORUTO  FOR PRESUMPTIVE DIAGNOSIS.
The above steps were performed and no features were selected.


```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
from sklearn.impute import SimpleImputer

# Load the dataset
df = pd.read_csv('/content/app_data_updated .csv')

# Encode Diagnosis_Presumptive (appendicitis vs. other)
# Ensure that non-string values are handled gracefully
df['Diagnosis_Presumptive'] = df['Diagnosis_Presumptive'].apply(
    lambda x: 1 if isinstance(x, str) and 'appendicitis' in x.lower() else 0
)
y_diagnosis = df['Diagnosis_Presumptive']

# Select features (exclude Diagnosis_Presumptive and Diagnosis)
X = df.drop(['Diagnosis_Presumptive', 'Diagnosis'], axis=1)

# Identify categorical and numerical columns
categorical_cols = X.select_dtypes(include=['object']).columns
numerical_cols = X.select_dtypes(include=['float64', 'int64']).columns

# Drop columns with >80% missing values
missing_threshold = 0.8
X = X.loc[:, X.isnull().mean() < missing_threshold]

# Update categorical and numerical columns after dropping
categorical_cols = [col for col in categorical_cols if col in X.columns]
numerical_cols = [col for col in numerical_cols if col in X.columns]

# Impute missing values
# Numerical: median
num_imputer = SimpleImputer(strategy='median')
X[numerical_cols] = num_imputer.fit_transform(X[numerical_cols])

# Categorical: mode
cat_imputer = SimpleImputer(strategy='most_frequent')
X[categorical_cols] = cat_imputer.fit_transform(X[categorical_cols])

# One-hot encode categorical variables
X = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

# Initialize Random Forest Classifier for Boruta
# Adjusted parameters to potentially select more features
rf = RandomForestClassifier(n_jobs=-1, class_weight='balanced', max_depth=None, n_estimators=1000, random_state=42)

# Run Boruta for Diagnosis_Presumptive
boruta_diagnosis = BorutaPy(rf, n_estimators='auto', verbose=2, random_state=42)
boruta_diagnosis.fit(X.values, y_diagnosis.values)

# Get selected features for Diagnosis_Presumptive
selected_features_diagnosis = X.columns[boruta_diagnosis.support_].tolist()

print("Selected features for Diagnosis_Presumptive:", selected_features_diagnosis)

# Save selected features to a CSV file
selected_features_diagnosis_df = pd.DataFrame(selected_features_diagnosis, columns=['Selected Features (Diagnosis_Presumptive)'])
selected_features_diagnosis_df.to_csv('selected_features_diagnosis_presumptive.csv', index=False)

print("Selected features for Diagnosis_Presumptive saved to 'selected_features_diagnosis_presumptive.csv'")
```
